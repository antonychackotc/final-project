```python
df1['Claim_History'].value_counts()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define claim history data from the provided counts
claim_history_data = {
    'Claim_History': [5, 3, 2, 1, 4, 0],
    'Count': [194, 181, 167, 161, 155, 142]
}

# Convert to DataFrame
df_claim = pd.DataFrame(claim_history_data)

# Calculate total count and percentage
total = df_claim['Count'].sum()
df_claim['Percentage'] = (df_claim['Count'] / total) * 100

# Assign rank based on count (descending order)
df_claim = df_claim.sort_values(by='Count', ascending=False).reset_index(drop=True)
df_claim['Rank'] = df_claim.index + 1

# Set color palette
colors = sns.color_palette("tab10", len(df_claim))

# Create a bar plot
plt.figure(figsize=(10, 6))
ax = sns.barplot(
    x=df_claim['Claim_History'].astype(str),  # Convert to string for proper labeling
    y=df_claim['Count'],
    palette=colors
)

# Annotate bars with percentage and rank
for index, row in df_claim.iterrows():
    ax.text(index, row['Count'] + total * 0.01,
            f"Rank {row['Rank']}\n{row['Percentage']:.1f}%",
            ha='center', fontsize=12, fontweight='bold')

# Add labels and title
plt.xlabel('Claim History')
plt.ylabel('Count')
plt.title('Distribution of Claim History with Rank & Percentage')

# Create legend to represent colors
legend_labels = [f"Rank {row['Rank']}: {row['Claim_History']}" for _, row in df_claim.iterrows()]
legend_patches = [plt.Rectangle((0, 0), 1, 1, color=colors[i]) for i in range(len(legend_labels))]
plt.legend(legend_patches, legend_labels, title="Claim History Ranks", loc="upper right", bbox_to_anchor=(1.3, 1))

# Show the plot
plt.show()
```

# **1) Plicy_Type vs Vehicle_Age_Property_Age**

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Define manual bins and labels
bins = [0, 5, 10, 15, 20, 25, 30, 35]
bin_labels = ['0-5', '6-10', '11-15', '16-20', '21-25', '26-30', '31-35']

# Create a new column for binned vehicle age
df1['Vehicle_Age_Bin'] = pd.cut(df1['Vehicle_Age_Property_Age'], bins=bins, labels=bin_labels, right=False)

# Calculate total count per bin for percentage calculation
bin_counts = df1['Vehicle_Age_Bin'].value_counts(normalize=False).to_dict()
bin_total = df1['Vehicle_Age_Bin'].value_counts(normalize=True).to_dict()

# Plot count distribution
plt.figure(figsize=(14, 6))
ax = sns.countplot(x='Vehicle_Age_Bin', hue='Policy_Type', data=df1, palette="husl")

# Annotate bars with count and percentage
for p in ax.patches:
    height = p.get_height()
    if height > 0:
        bin_label = p.get_x() + p.get_width() / 2  # Get bin label position
        total_count = bin_counts.get(ax.get_xticklabels()[int(p.get_x())].get_text(), 1)
        percentage = (height / total_count) * 100  # Compute percentage

        # Display count and percentage
        ax.annotate(f"{int(height)}\n({percentage:.1f}%)",
                    (bin_label, height),
                    ha='center', va='bottom', fontsize=10, fontweight='bold')

# Set labels and title
plt.xlabel('Vehicle Age Groups')
plt.ylabel('Count')
plt.title('Policy Type Distribution Across Vehicle Age Groups')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Add legend
plt.legend(title="Policy Type", loc="upper right")

# Show the plot
plt.show()
```

# **6) Customer_Age_Bin vs	Premium_Amount, Claim_Amount**

**Claim amounts show a much wider distribution and greater variability compared to premiums.**

**Average Claim Amount (Red Line):**


The **average claim amount fluctuates more than the premium**. There are **peaks in the 46-55 and 56-65 age groups**, aligning with the boxplot observations.
